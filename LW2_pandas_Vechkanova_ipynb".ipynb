7. Создание нового поля вычисленного на основе значений других полей:

7.1. Через выражение на базе имеющихся колонок.

7.2. Через `DataFrame.apply`

7.3. Через `Series.apply`


```python
df["petal square (cm^2)"] = df["petal length (cm)"]*df["petal width (cm)"]
df
```

```python
def sepal_square (row):
  return row["sepal length (cm)"]*row["sepal width (cm)"]

df["sepal square (cm^2)"] = df.apply(sepal_square, axis=1)
df
```

```python
sep_len = df['sepal length (cm)']
pet_len = df['petal length (cm)']

total_length = pd.Series(sep_len + pet_len)

df['total length (m)'] = total_length.apply(lambda row: row/100)
df
```

##Задание 3
Продемонстрировать работу `.groupby`, на основе группировок в groupby вычисляете агрегатные функции по одной или нескольким колонкам

```python
d = df.groupby("total length (m)").agg({'sepal length (cm)': 'mean', 'petal length (cm)': 'sum'})
d
```

```python
e = d.reset_index().pivot(index = "petal length (cm)", columns = "sepal length (cm)",	values = "total length (m)")
e
```
